```python
import sys
sys.path.append(".secrets")

from google.oauth2.service_account import Credentials
from langchain_google_vertexai import VertexAIEmbeddings
from load_credentials import load_credentials
import numpy as np
```

```python
def get_embedding(text: str, credentials) -> list[float]:
    """
    Takes a text and returns its embedding as a Python list using Google's Gemini embedding model via Vertex AI.
    """

    embeddings_model = VertexAIEmbeddings(
        model_name="gemini-embedding-001",
        credentials=credentials,
    )
    return embeddings_model.embed_query(text)


def cosine_similarity(embedding_a: list[float], embedding_b: list[float]) -> float:
    a = np.array(embedding_a)
    b = np.array(embedding_b)
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
```

```python
creds = load_credentials()

emb_madre = get_embedding("Madre", creds)
emb_mujer = get_embedding("Mujer", creds)
emb_hombre = get_embedding("Hombre", creds)
emb_padre = get_embedding("Padre", creds)

emb_padre_calcuado = [madre - mujer + hombre for madre, mujer, hombre in zip(emb_madre, emb_mujer, emb_hombre)]

cosine_similarity(emb_padre_calcuado, emb_padre)
```
